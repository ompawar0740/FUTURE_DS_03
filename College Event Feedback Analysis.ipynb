```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import re
from collections import Counter
import numpy as np

# Set plotting style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Create a dummy CSV file for demonstration if you don't have one
data = {
    'Timestamp': pd.to_datetime(['2025-07-20 10:00:00', '2025-07-20 10:15:00', '2025-07-20 10:30:00', '2025-07-20 11:00:00', '2025-07-20 11:30:00', '2025-07-20 12:00:00', '2025-07-20 12:30:00', '2025-07-20 13:00:00', '2025-07-20 13:30:00', '2025-07-20 14:00:00']),
    'Overall_Satisfaction': [5, 4, 3, 5, 4, 2, 5, 4, 3, 5],
    'Speaker_Quality': [4, 5, 3, 4, 4, 2, 5, 3, 4, 5],
    'Content_Relevance': [5, 4, 4, 5, 3, 3, 4, 5, 3, 4],
    'Venue_Logistics': [4, 4, 3, 5, 4, 2, 4, 4, 3, 5],
    'Attend_Again': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'Liked_Most': [
        'Great speakers and engaging content!',
        'Networking opportunities were excellent.',
        'The venue was a bit crowded.',
        'Loved the interactive sessions.',
        'Good variety of topics.',
        'Too long and some speakers were boring.',
        'Everything was perfectly organized.',
        'Learned a lot from the workshops.',
        'Food could be better.',
        'Enthusiastic organizers.'
    ],
    'Suggestions': [
        'More breaks between sessions.',
        'No suggestions, it was perfect.',
        'Improve the sound system.',
        'Include a Q&A panel with all speakers.',
        'Offer more advanced topics next time.',
        'Shorter sessions, less theory.',
        'None.',
        'Provide clearer directions to the breakout rooms.',
        'Better food options.',
        'Keep up the great work!'
    ],
    'Year_of_Study': ['First Year', 'Second Year', 'Third Year', 'Fourth Year', 'First Year', 'Second Year', 'Third Year', 'Fourth Year', 'First Year', 'Second Year']
}
df = pd.DataFrame(data)
df.to_csv('event_feedback.csv', index=False)

# Load the dataset
try:
    df = pd.read_csv('event_feedback.csv')
    print("Data loaded successfully!")
    print(df.head())
except FileNotFoundError:
    print("Error: 'event_feedback.csv' not found. Please ensure the file is in the correct directory.")

print("\n--- Data Info ---")
df.info()

print("\n--- Missing Values ---")
print(df.isnull().sum())

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Attend_Again'] = df['Attend_Again'].replace({'yes': 'Yes', 'no': 'No'})

print("\n--- Descriptive Statistics for Ratings ---")
rating_columns = ['Overall_Satisfaction', 'Speaker_Quality', 'Content_Relevance', 'Venue_Logistics']
print(df[rating_columns].describe())

print("\n--- 'Would you attend again?' Distribution ---")
print(df['Attend_Again'].value_counts(normalize=True))

plt.figure(figsize=(12, 6))
for i, col in enumerate(rating_columns):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'Distribution of {col.replace("_", " ")}')
    plt.xlabel('Rating (1-5)')
    plt.ylabel('Number of Responses')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Year_of_Study', y='Overall_Satisfaction', palette='coolwarm', order=['First Year', 'Second Year', 'Third Year', 'Fourth Year'])
plt.title('Overall Satisfaction by Year of Study')
plt.xlabel('Year of Study')
plt.ylabel('Overall Satisfaction Rating')
plt.show()

attend_counts = df['Attend_Again'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(attend_counts, labels=attend_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Likelihood to Attend Again')
plt.show()

def get_sentiment(text):
    if pd.isna(text):
        return None
    return TextBlob(text).sentiment.polarity

df['Liked_Most_Sentiment'] = df['Liked_Most'].apply(get_sentiment)
df['Suggestions_Sentiment'] = df['Suggestions'].apply(get_sentiment)

print("\n--- Sentiment Analysis Summary for 'Liked Most' ---")
print(df['Liked_Most_Sentiment'].describe())

print("\n--- Sentiment Analysis Summary for 'Suggestions' ---")
print(df['Suggestions_Sentiment'].describe())

def categorize_sentiment(polarity):
    if polarity is None:
        return 'N/A'
    elif polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['Liked_Most_Sentiment_Category'] = df['Liked_Most_Sentiment'].apply(categorize_sentiment)
df['Suggestions_Sentiment_Category'] = df['Suggestions_Sentiment'].apply(categorize_sentiment)

print("\n--- Sentiment Category Distribution for 'Liked Most' ---")
print(df['Liked_Most_Sentiment_Category'].value_counts(normalize=True))

print("\n--- Sentiment Category Distribution for 'Suggestions' ---")
print(df['Suggestions_Sentiment_Category'].value_counts(normalize=True))

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.countplot(data=df, x='Liked_Most_Sentiment_Category', palette='pastel', ax=axes[0], order=['Positive', 'Neutral', 'Negative', 'N/A'])
axes[0].set_title('Sentiment Distribution for "What was liked most?"')
axes[0].set_xlabel('Sentiment Category')
axes[0].set_ylabel('Number of Responses')

sns.countplot(data=df, x='Suggestions_Sentiment_Category', palette='pastel', ax=axes[1], order=['Positive', 'Neutral', 'Negative', 'N/A'])
axes[1].set_title('Sentiment Distribution for "Suggestions for improvement"')
axes[1].set_xlabel('Sentiment Category')
axes[1].set_ylabel('Number of Responses')
plt.tight_layout()
plt.show()

def clean_text_for_wordcloud(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['Liked_Most_Cleaned'] = df['Liked_Most'].apply(clean_text_for_wordcloud)
df['Suggestions_Cleaned'] = df['Suggestions'].apply(clean_text_for_wordcloud)

all_liked_words = " ".join(df['Liked_Most_Cleaned'].dropna()).split()
all_suggestion_words = " ".join(df['Suggestions_Cleaned'].dropna()).split()

stop_words = set([
    "the", "a", "an", "is", "it", "to", "and", "of", "in", "for", "with", "was", "were", "on", "at", "from",
    "that", "this", "be", "have", "i", "you", "he", "she", "we", "they", "not", "but", "very", "good", "great",
    "excellent", "well", "could", "would", "should", "some", "most", "also", "just", "much", "more", "less", "no", "none"
])

filtered_liked_words = [word for word in all_liked_words if word not in stop_words and len(word) > 2]
filtered_suggestion_words = [word for word in all_suggestion_words if word not in stop_words and len(word) > 2]

common_liked_words = Counter(filtered_liked_words).most_common(15)
common_suggestion_words = Counter(filtered_suggestion_words).most_common(15)

print("\n--- Top 15 Most Liked Keywords ---")
print(common_liked_words)

print("\n--- Top 15 Suggestion Keywords ---")
print(common_suggestion_words)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

liked_words_df = pd.DataFrame(common_liked_words, columns=['Word', 'Count'])
sns.barplot(x='Count', y='Word', data=liked_words_df, palette='GnBu_r', ax=axes[0])
axes[0].set_title('Top Keywords from "What was liked most?"')
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('Keyword')

suggestion_words_df = pd.DataFrame(common_suggestion_words, columns=['Word', 'Count'])
sns.barplot(x='Count', y='Word', data=suggestion_words_df, palette='OrRd_r', ax=axes[1])
axes[1].set_title('Top Keywords from "Suggestions for improvement"')
axes[1].set_xlabel('Frequency')
axes[1].set_ylabel('Keyword')

plt.tight_layout()
plt.show()

try:
    from wordcloud import WordCloud

    wordcloud_liked = WordCloud(width=800, height=400, background_color='white').generate(" ".join(filtered_liked_words))
    wordcloud_suggestions = WordCloud(width=800, height=400, background_color='white').generate(" ".join(filtered_suggestion_words))

    plt.figure(figsize=(15, 7))
    plt.subplot(1, 2, 1)
    plt.imshow(wordcloud_liked, interpolation='bilinear')
    plt.title('Word Cloud: What was Liked Most?')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(wordcloud_suggestions, interpolation='bilinear')
    plt.title('Word Cloud: Suggestions for Improvement')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

except ImportError:
    print("\n'wordcloud' library not found. Install with: pip install wordcloud for word cloud visualizations.")

print("\n--- Correlation Matrix for Ratings ---")
correlation_matrix = df[rating_columns].corr()
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Event Ratings')
plt.show()
```
